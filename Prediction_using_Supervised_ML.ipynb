Importing all the required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

Importing the dataset

```python
data = pd.read_csv('http://bit.ly/w-data')
data.info()
```

No null values


```python
data.shape
```

```python
data.duplicated().sum()
```

No duplicate values


```python
data.describe()
```

* **Data Visualization**





```python
data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

* **Preparing the data and splitting it into test and train data**

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values

trainx, testx, trainy, testy = train_test_split(X, Y, test_size=0.2, random_state = 1)
```

* **Training, testing and predicting the model**

```python
reg = LinearRegression()
reg.fit(trainx, trainy)
predy = reg.predict(testx)
```

* **Comparing actual with predicted**

```python
df = pd.DataFrame({'Actual':testy, 'Predicted':predy})
df
```

* **Accuracy of the model**

```python
metrics.r2_score(testy,predy)
```

* **Evaluating the model**

```python
print("Mean Squared Error  > ",metrics.mean_squared_error(testy,predy))
print("Mean Absolute Error > ",mean_absolute_error(testy,predy))
```

# **Predicted score if a student studies for 9.25 hrs/day**

```python
hours = [9.25]
result = reg.predict([hours])
print("Marks = {}".format(round(result[0],8)))
```
